###Python Generators
Performs lazy evaluation and do not generate data unless needed. Can generate infinite numbers without storing them in memory.

````
def factors(n):
  k = 1
  while k * k < n:
    if n % k == 0:
      yield k
      yield n // k
    k += 1
  if k * k == n: # Special case to handle perfect squares numbers
    yield k
````
